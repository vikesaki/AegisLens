```python
# import libraries
import cv2
import os
import shutil
```

```python
# num list
num_list = ['06', '07']

# take frame per how many frames
frame_interval = 120

for num in num_list:
    # assign path
    name = 'ojan'+ num
    video_path = rf'raw images\raw videos\{name}.MOV'
    output_dir = rf'raw images\{name}'

    # create output directory
    os.makedirs(output_dir, exist_ok=True)

    # open video
    cap = cv2.VideoCapture(video_path)

    # if video not opened
    if not cap.isOpened():
        print("Error: Cannot open video.")
        exit()

    # count frame
    frame_count = 0
    saved_count = 0

    # run program
    while True:
        # run video and parse frame
        ret, frame = cap.read()

        # if no frame returned
        if not ret:
            break
        
        # if frame count and frame interval modulus 0
        if frame_count % frame_interval == 0:
            # name frame
            frame_name = f"{name}_{saved_count:05d}.jpg"

            # path for frame
            save_path = os.path.join(output_dir, frame_name)

            # save frame
            cv2.imwrite(save_path, frame)

            # print output
            print(f"Saved: {frame_name}")

            # increase saved frame count
            saved_count += 1

        # increase frame count
        frame_count += 1

    # stop video
    cap.release()

    # output
    print(f"\nDone! Saved {saved_count} frames to '{output_dir}'.")
```

```python
num_list = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11']

# Destination folder where all files will be moved
destination_folder = r'raw images\images'

# Create destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Loop through each folder and move files
for num in num_list:
    name = 'ojan'+ num
    folder = rf'raw images\{name}'
    
    for filename in os.listdir(folder):
        src_file = os.path.join(folder, filename)
        dest_file = os.path.join(destination_folder, filename)

        # Check if it's a file (not a subfolder)
        if os.path.isfile(src_file):
            # Optional: handle duplicate filenames
            if os.path.exists(dest_file):
                base, ext = os.path.splitext(filename)
                i = 1
                while os.path.exists(dest_file):
                    dest_file = os.path.join(destination_folder, f"{base}_{i}{ext}")
                    i += 1
            
            shutil.move(src_file, dest_file)
            print(f"Moved: {src_file} → {dest_file}")
```

```python
# Define valid image extensions
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff', '.webp')

# List and sort the files for consistent ordering
files = sorted([f for f in os.listdir(destination_folder) if f.lower().endswith(image_extensions)])

# Start renaming
for idx, filename in enumerate(files, start=1):
    ext = os.path.splitext(filename)[1]  # get file extension
    new_name = f'image_{idx:05d}{ext}'   # zero-padded numbering
    src = os.path.join(destination_folder, filename)
    dst = os.path.join(destination_folder, new_name)

    os.rename(src, dst)
    print(f'Renamed: {filename} → {new_name}')
```
